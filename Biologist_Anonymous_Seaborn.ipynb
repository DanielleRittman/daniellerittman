```python
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
kidsdata = pd.read_sas('kidsdata.sas7bdat')
```

```python
sns.set()
columns_wanted = ['telomere_ts', 'i_mets_z', 'i_male']
important_column = 'i_male' #column facetted on 
#from kidsdata gather necessary data then facet on the important column
g = sns.FacetGrid(kidsdata.loc[:,columns_wanted], col=important_column, hue=important_column, col_wrap=2, legend_out = True)
sns_plot = (g.map(plt.scatter, 'telomere_ts', 'i_mets_z').set_axis_labels("Telomere Length", "Metabolic Score")
            .set_titles("").add_legend())#graphs as scatter plot and renames the axis, adds legend
new_title = 'Gender'
g._legend.set_title(new_title) #renames legend
# replace labels
new_labels = ['Female', 'Male']
for t, l in zip(g._legend.texts, new_labels): t.set_text(l)#set the labels in the legend
sns_plot.savefig("mets_gender")
```
